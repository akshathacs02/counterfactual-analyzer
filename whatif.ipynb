```python
from openai import OpenAI

# Initialize client
client = OpenAI(api_key="your_api")

# User inputs
scenario = input("Enter the original scenario:\n")
change = input("\nEnter the 'what-if' change:\n")

# Prompt
prompt = f"""
You are a Counterfactual Analyzer.

Original Scenario:
{scenario}

What-if Change:
{change}

Analyze:
1. Possible impacts
2. Positive effects
3. Negative effects
4. Final reasoning
"""

# API call
response = client.chat.completions.create(
    model="gpt-5-mini",
    messages=[
        {"role": "user", "content": prompt}
    ]
)

# Output
print("\n=== Counterfactual Analysis ===\n")
print(response.choices[0].message.content)
```

With UI integration

```python
# Install dependencies (run in Colab if needed)
!pip install openai gradio -q

from openai import OpenAI
import gradio as gr

# SET YOUR API KEY
client = OpenAI(api_key="YOUR_API_KEY")

# CORE FUNCTION
def counterfactual_analyzer(scenario, change):

    prompt = f"""
You are a Counterfactual Analyzer.

Original Scenario:
{scenario}

What-if Change:
{change}

Provide:
1. Possible impacts
2. Positive effects
3. Negative effects
4. Reasoned conclusion
"""

    response = client.chat.completions.create(
        model="gpt-5-mini",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content

# GRADIO UI
with gr.Blocks() as app:
    gr.Markdown("# 🔍 Counterfactual Analyzer")
    gr.Markdown("Explore what might happen if one key factor changes in a scenario.")

    with gr.Row():
        scenario_input = gr.Textbox(
            label="Original Scenario",
            placeholder="e.g., A company increased prices and sales dropped"
        )

    change_input = gr.Textbox(
        label="What-if Change",
        placeholder="e.g., What if prices were reduced instead?"
    )

    output = gr.Textbox(label="Analysis Result", lines=15)

    btn = gr.Button("Analyze")

    btn.click(
        fn=counterfactual_analyzer,
        inputs=[scenario_input, change_input],
        outputs=output
    )

# LAUNCH APP
app.launch()
```
